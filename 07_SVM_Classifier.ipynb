# 🎯 SVM Classifier — Quick Notes
---
> **Type:** Supervised | **Task:** Classification | **Dataset:** `Social_Network_Ads`

## 📌 Key Points
- Finds the **best hyperplane** that separates classes
- Maximizes the **margin** between classes
- Points on boundary = **Support Vectors**
- **Kernel trick** → handles non-linear data
- **Must scale** features!
- Slow on large datasets

## 🔑 Kernels
| Kernel | Use When |
|---|---|
| `linear` | Linearly separable data |
| `rbf` (default) | Non-linear, most common |
| `poly` | Polynomial boundary |

## 📊 Dataset: `Social_Network_Ads`
| Column | Role |
|---|---|
| `Gender`, `Age`, `EstimatedSalary` | Features (X) |
| `Purchased` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv("/content/Social_Network_Ads - Social_Network_Ads.csv")
lb = LabelEncoder()
df['Gender'] = lb.fit_transform(df['Gender'])

x = df.drop(columns=['Purchased'])
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test  = sc.transform(x_test)

svm = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## 🔧 Key Parameters
| Param | Meaning |
|---|---|
| `C` | Regularization — high C = fits more |
| `gamma` | Influence of each point |
| `kernel` | Shape of decision boundary |

## 🥇 Remember
- **Always scale** before SVM
- `C` high → overfit | `C` low → underfit
- `probability=True` to get `predict_proba()`
