# ⚡ XGBoost Regressor — Quick Notes
---
> **Type:** Supervised | **Task:** Regression | **Dataset:** `insurance`

## 📌 Key Points
- XGBoost for continuous target prediction
- Same advantages as XGBoost Classifier
- Often **best regression model** on structured data
- `eval_metric='rmse'` for regression (not logloss)

## 📊 Dataset: `insurance`
| Column | Role |
|---|---|
| `age`, `bmi`, `children`, `sex`, `smoker`, `region` | Features (X) |
| `charges` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error

df = pd.read_csv("/content/insurance - insurance.csv")
lb = LabelEncoder()
for col in ['sex','smoker','region']:
    df[col] = lb.fit_transform(df[col])

x = df.drop(columns=['charges'])
y = df['charges']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3,
                   eval_metric='rmse', random_state=42)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE     :", mean_absolute_error(y_test, y_pred))
```

## 🥇 Remember
- Switch `eval_metric='rmse'` for regression
- Top performer on insurance-type structured data
- `XGBRegressor` vs `XGBClassifier` — only class name & metric differ
