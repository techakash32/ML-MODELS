# 🎯 SVR — Support Vector Regressor — Quick Notes
---
> **Type:** Supervised | **Task:** Regression | **Dataset:** `insurance`

## 📌 Key Points
- Regression version of SVM
- Fits data within an **epsilon (ε) tube** — no penalty inside
- Points outside tube = support vectors
- **Must scale both X and y**
- Kernel trick works same as SVM

## 📊 Dataset: `insurance`
| Column | Role |
|---|---|
| `age`, `bmi`, `children`, `sex`, `smoker`, `region` | Features (X) |
| `charges` | Target (y) |

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error

df = pd.read_csv("/content/insurance - insurance.csv")
lb = LabelEncoder()
for col in ['sex','smoker','region']:
    df[col] = lb.fit_transform(df[col])

x = df.drop(columns=['charges'])
y = df['charges']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test  = sc_x.transform(x_test)
y_train_scaled = sc_y.fit_transform(y_train.values.reshape(-1,1)).ravel()

svr = SVR(kernel='rbf', C=100, epsilon=0.1)
svr.fit(x_train, y_train_scaled)
y_pred = sc_y.inverse_transform(svr.predict(x_test).reshape(-1,1)).ravel()

print("R2 Score:", r2_score(y_test, y_pred))
print("MAE     :", mean_absolute_error(y_test, y_pred))
```

## 🥇 Remember
- Scale **both X and y** for SVR
- `epsilon` → size of no-penalty tube
- Always inverse transform y_pred after prediction
