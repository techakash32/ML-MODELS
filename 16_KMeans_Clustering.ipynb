# 🔵 K-Means Clustering — Quick Notes
---
> **Type:** Unsupervised | **Task:** Clustering | **Dataset:** `mall`

## 📌 Key Points
- Groups data into **K clusters** by minimizing within-cluster distance
- Uses **Elbow Method** to find best K (look for the "elbow" in WCSS plot)
- `init='k-means++'` → smarter centroid initialization (avoids random bad starts)
- Sensitive to **outliers** and **feature scale**
- Must specify K in advance
- Iterates: assign points → update centroids → repeat until stable

## 📊 Dataset: `mall`
| Column | Role |
|---|---|
| `Age`, `Annual Income (k$)` | Features (x[:,0], x[:,1]) |
| `Spending Score (1-100)` | Can also be used as 3rd feature |

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. Load & Prepare
df = pd.read_csv("/content/mall - mall.csv")
df = df.drop(columns=['CustomerID', 'Genre'])
x = df.iloc[:, [0, 1]].values   # Age vs Annual Income

# 2. Elbow Method — find best K
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)

plt.figure(figsize=(8,4))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method — Find Best K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.show()
# Look for the elbow point → that's your K

# 3. Train with best K = 4
km = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_pred = km.fit_predict(x)

# 4. Visualize Clusters
colors = ['blue','green','red','cyan']
labels = [f'Cluster {i+1}' for i in range(4)]
plt.figure(figsize=(8,6))
for i in range(4):
    plt.scatter(x[y_pred==i, 0], x[y_pred==i, 1], s=80, c=colors[i], label=labels[i])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],
            s=300, c='yellow', marker='*', label='Centroids')
plt.title('K-Means Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.legend()
plt.show()

print("Inertia (WCSS):", km.inertia_)
print("Cluster Centers:\n", km.cluster_centers_)
```

## 🔧 Key Parameters
| Param | Meaning |
|---|---|
| `n_clusters` | Number of clusters K |
| `init` | `k-means++` (smart) or `random` |
| `max_iter` | Max iterations per run |
| `n_init` | Times algorithm runs with different centroids |

## 🥇 Remember
- **Elbow point** = where WCSS stops dropping sharply → best K
- `inertia_` = total within-cluster sum of squares (WCSS)
- Scale features with StandardScaler if ranges differ a lot
- Works best on **spherical / blob-shaped** clusters
