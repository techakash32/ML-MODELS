```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/credit_scoring - credit_scoring.csv")
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['Gender']=lb.fit_transform(df['Gender'])
df['Marital Status']=lb.fit_transform(df['Marital Status'])
df['Education Level']=lb.fit_transform(df['Education Level'])
df['Employment Status']=lb.fit_transform(df['Employment Status'])
df['Type of Loan']=lb.fit_transform(df['Type of Loan'])
```

```python
x=df.drop(columns=['Loan Amount'])
y=df['Loan Amount']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn. linear_model import LinearRegression
```

```python
lr=LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```
